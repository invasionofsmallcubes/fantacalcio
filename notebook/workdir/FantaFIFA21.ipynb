```python
import pandas as pd
import re
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 500)
```

```python
def convertAndSum(x):
    listOfStrings = x.split('+')
    lst = [int(i) if i != '' else 0 for i in listOfStrings]
    return sum(lst)

def cleanName(x):
    a = re.sub(r'[A-Za-z]\.', '', x)
    return a.lower()
```

```python
data = pd.read_csv (r'players_21.csv')
```

```python
data.drop(columns=['id', 'value_eur','nation_jersey_number','contract_valid_until', 'nation_position','loaned_from','joined','team_position','work_rate', 'wage_eur', 'international_reputation','player_url', 'nationality','release_clause_eur','team_jersey_number', 'long_name', 'date_of_birth', 'height_cm', 'weight_kg', 'league_name', 'league_rank', 'body_type', 'real_face'], inplace=True)
```

```python
data['ldm'] = data['ldm'].apply(lambda x: convertAndSum(x))
data['cdm'] = data['cdm'].apply(lambda x: convertAndSum(x))
data['rdm'] = data['rdm'].apply(lambda x: convertAndSum(x))
data['rwb'] = data['rwb'].apply(lambda x: convertAndSum(x))
data['lb'] = data['lb'].apply(lambda x: convertAndSum(x))
data['lcb'] = data['lcb'].apply(lambda x: convertAndSum(x))
data['cb'] = data['cb'].apply(lambda x: convertAndSum(x))
data['rcb'] = data['rcb'].apply(lambda x: convertAndSum(x))
data['rb'] = data['rb'].apply(lambda x: convertAndSum(x))
data['ls'] = data['ls'].apply(lambda x: convertAndSum(x))
data['st'] = data['st'].apply(lambda x: convertAndSum(x))
data['rs'] = data['rs'].apply(lambda x: convertAndSum(x))
data['lw'] = data['lw'].apply(lambda x: convertAndSum(x))
data['lf'] = data['lf'].apply(lambda x: convertAndSum(x))
data['cf'] = data['cf'].apply(lambda x: convertAndSum(x))
data['rf'] = data['rf'].apply(lambda x: convertAndSum(x))
data['rw'] = data['rw'].apply(lambda x: convertAndSum(x))
data['lam'] = data['lam'].apply(lambda x: convertAndSum(x))
data['cam'] = data['cam'].apply(lambda x: convertAndSum(x))
data['ram'] = data['ram'].apply(lambda x: convertAndSum(x))
data['lm'] = data['lm'].apply(lambda x: convertAndSum(x))
data['lcm'] = data['lcm'].apply(lambda x: convertAndSum(x))
data['cm'] = data['cm'].apply(lambda x: convertAndSum(x))
data['rcm'] = data['rcm'].apply(lambda x: convertAndSum(x))
data['rm'] = data['rm'].apply(lambda x: convertAndSum(x))
data['lwb'] = data['lwb'].apply(lambda x: convertAndSum(x))
```

```python
data['short_name'] = data['short_name'].apply(lambda x: cleanName(x))
data.short_name = data.short_name.astype(str)
data.short_name = data.short_name.str.strip()
```

```python
compr = pd.read_csv (r'giocabili.csv', delimiter='\t')
```

```python
compr['short_name'] = compr['short_name'].apply(lambda x: cleanName(x))
compr.short_name = compr.short_name.astype(str)
compr.short_name = compr.short_name.str.strip()
compr = compr.drop(['DRR', 'RUOLO'], axis=1)
compr = compr.fillna(0)
compr.OVERA = compr.OVERA.astype(int)
compr.ETA = compr.ETA.astype(int)
```

```python
onTheMarketInternal = onTheMarket.merge(compr, how='right', on='short_name')
onTheMarketInternal = onTheMarketInternal.drop(columns=['age'])
```

```python
onTheMarketExternal = onTheMarket.merge(compr, how='left', on='short_name')
onTheMarketExternal = onTheMarketExternal[onTheMarketExternal['OVERA'].isna()]
```
